```python

# 📁 Imports
from pydub import AudioSegment
from pyannote.audio import Pipeline
import whisper
import torch
import json
import os
from datetime import datetime


token = "XXXXXXXXXXXXXXXXXXXXXXXX"
#Remplacer

# 📍 Paramètres généraux
audio_path = "/content/drive/MyDrive/AudiosToTranscribe/Archive_007.mp3"  # à remplacer
archive_title = "ARCHIVE_007"
mode_test = False
```

```python

# 🎧 Conversion en WAV mono 16kHz
audio = AudioSegment.from_mp3(audio_path)
wav = audio.set_channels(1).set_frame_rate(16000)

# 🎯 Mode test : extrait de 1:15 à 1:45 (30s)
if mode_test:
    start_ms = 75 * 1000
    end_ms = 105 * 1000
    clip = wav[start_ms:end_ms]
    wav_path = "test_clip.wav"
else:
    wav_path = "converted.wav"
    clip = wav

clip.export(wav_path, format="wav")
print(f"✅ Fichier audio exporté : {wav_path}")
```

```python
from faster_whisper import WhisperModel

model = WhisperModel("large-v3", device="cuda", compute_type="float16")
segments, info = model.transcribe(wav_path, word_timestamps=True)
segments = list(segments)

print(f"✅ {len(segments)} segments transcrits")
```

```python
# 🧩 Construction du JSON avec découpage naturel (ponctuation) et max 7 mots
final_data = {
    "title": archive_title,
    "instrumentals": [
        {"start": 0.0, "end": 60.0, "title": "PLACEHOLDER 1", "artist": "TODO"},
        {"start": 60.0, "end": 120.0, "title": "PLACEHOLDER 2", "artist": "TODO"},
        {"start": 120.0, "end": 180.0, "title": "PLACEHOLDER 3", "artist": "TODO"}
    ],
    "segments": []
}

def find_speaker(start, end):
    for s in speaker_timeline:
        if s["start"] <= start < s["end"]:
            return s["speaker"]
    return "unknown"

def split_by_diction(words, max_words=7):
    chunks = []
    current = []
    for word in words:
        w = word["word"].strip() if isinstance(word, dict) else word.word.strip()
        current.append(word)
        if len(current) >= max_words or w.endswith((".", "!", "?", ",", ";", "…")):
            chunks.append(current)
            current = []
    if current:
        chunks.append(current)
    return chunks

segment_id = 0
for seg in segments:
    words_raw = seg.words if hasattr(seg, "words") else seg["words"]

    word_dicts = []
    for word in words_raw:
        word = word._asdict() if hasattr(word, "_asdict") else word
        word.pop("seek", None)
        word.pop("probability", None)
        word_dicts.append(word)

    # découpage par diction + limite 7 mots
    chunks = split_by_diction(word_dicts, max_words=7)

    for chunk in chunks:
        final_data["segments"].append({
            "id": segment_id,
            "start": chunk[0]["start"],
            "end": chunk[-1]["end"],
            "text": " ".join([w["word"] for w in chunk]),
            "words": chunk,
            "speaker": find_speaker(chunk[0]["start"], chunk[-1]["end"]),
            "instrumental": None
        })
        segment_id += 1

# Export
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
filename = f"archive_output_{timestamp}.json"
with open(filename, "w") as f:
    json.dump(final_data, f, indent=2)

print(f"✅ Fichier exporté : {filename}")
```
